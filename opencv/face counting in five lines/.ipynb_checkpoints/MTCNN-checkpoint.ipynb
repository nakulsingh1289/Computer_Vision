# Face Dectection with MTCNN

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# !pip install mtcnn
```

```python
from mtcnn import MTCNN
model = MTCNN()
img = cv2.imread("img.jpg")
result = model.detect_faces(img)
```

```python
result
```

```python
print("Total number of faces {}".format(len(result)))
```

```python
def show_faces(result):
    for i in range(len(result)):
        if result[i]["confidence"]>0.90:
            x1, y1, w, h= result[i]['box']
            x1, y1= abs(x1), abs(y1)
            x2, y2= x1+w, y1+h
            pic = cv2.rectangle(img, (x1,y1), (x2, y2), (0,255,255), 5)
    return pic
```

```python
pic = show_faces(result)
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.imshow(pic)
```
